```python
%%writefile /content/make_rtmdet_config.py
from mmengine.config import Config

BASE_CONFIG = "/content/mmdetection/configs/rtmdet/rtmdet_tiny_8xb32-300e_coco.py"
OUT_CONFIG = "/content/mmdetection/configs/rtmdet/rtmdet_tiny_sidewalk_hazards.py"

DATA_ROOT = "/content/MVProject/"
WORK_DIR = "/content/drive/MyDrive/MVProject/rtmdet_tiny_work_dir"
CHECKPOINT = "/content/drive/MyDrive/MVProject/checkpoints/rtmdet_tiny_coco.pth"

CLASSES = (
    "person", "bicycle", "car", "traffic_cone", "pothole", "open_manhole",
    "pole", "sign", "trash_can", "roadblock", "tree", "animal",
    "fire_hydrant", "crosswalk", "tactile_paving", "surface_damage",
)

cfg = Config.fromfile(BASE_CONFIG)

cfg.dataset_type = "CocoDataset"
cfg.data_root = DATA_ROOT
cfg.metainfo = dict(classes=CLASSES)

cfg.train_dataloader.dataset.type = "CocoDataset"
cfg.train_dataloader.dataset.data_root = DATA_ROOT
cfg.train_dataloader.dataset.ann_file = "annotations/train.json"
cfg.train_dataloader.dataset.data_prefix = dict(img="images/train/")
cfg.train_dataloader.dataset.metainfo = cfg.metainfo

cfg.val_dataloader.dataset.type = "CocoDataset"
cfg.val_dataloader.dataset.data_root = DATA_ROOT
cfg.val_dataloader.dataset.ann_file = "annotations/val.json"
cfg.val_dataloader.dataset.data_prefix = dict(img="images/val/")
cfg.val_dataloader.dataset.metainfo = cfg.metainfo

cfg.test_dataloader = cfg.val_dataloader
cfg.val_evaluator.ann_file = DATA_ROOT + "annotations/val.json"
cfg.test_evaluator = cfg.val_evaluator

cfg.model.bbox_head.num_classes = len(CLASSES)
cfg.load_from = CHECKPOINT

cfg.train_dataloader.batch_size = 2
cfg.train_dataloader.num_workers = 1
cfg.val_dataloader.batch_size = 2
cfg.val_dataloader.num_workers = 1
cfg.test_dataloader.batch_size = 2
cfg.test_dataloader.num_workers = 1

cfg.train_cfg.max_epochs = 10
cfg.train_cfg.val_interval = 1
cfg.optim_wrapper.optimizer.lr = 1e-4
cfg.optim_wrapper.type = "AmpOptimWrapper"

cfg.work_dir = WORK_DIR
cfg.default_hooks.checkpoint.interval = 1
cfg.default_hooks.checkpoint.max_keep_ckpts = 5
cfg.default_hooks.logger.interval = 50

for pipeline in [cfg.train_pipeline, cfg.test_pipeline]:
    for step in pipeline:
        if "scale" in step:
            step["scale"] = (512, 512)

cfg.dump(OUT_CONFIG)

print("Wrote config:", OUT_CONFIG)
print("Classes:", CLASSES)
print("num_classes:", len(CLASSES))
print("data_root:", DATA_ROOT)
print("work_dir:", WORK_DIR)
```

```python
%%writefile /content/test_model_metrics.py
import os
import csv
import time
import argparse
import cv2
import torch
from tqdm import tqdm
from mmdet.apis import init_detector, inference_detector


def parse_args():
    parser = argparse.ArgumentParser()
    parser.add_argument("--config", required=True)
    parser.add_argument("--checkpoint", required=True)
    parser.add_argument("--input", required=True, help="Path to image folder")
    parser.add_argument("--output", default="outputs")
    parser.add_argument("--device", default="cuda:0")
    parser.add_argument("--conf", type=float, default=0.35)
    return parser.parse_args()


def save_summary_csv(output_dir, metrics):
    path = os.path.join(output_dir, "model_test_metrics.csv")
    with open(path, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=metrics.keys())
        writer.writeheader()
        writer.writerow(metrics)
    print(f"Saved summary CSV to: {path}")


def save_per_image_csv(output_dir, rows):
    path = os.path.join(output_dir, "per_image_metrics.csv")
    if not rows:
        return
    with open(path, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=rows[0].keys())
        writer.writeheader()
        writer.writerows(rows)
    print(f"Saved per-image CSV to: {path}")


def get_direction(box, image_width):
    x1, _, x2, _ = box
    center_x = (x1 + x2) / 2
    if center_x < image_width / 3:
        return "left"
    elif center_x < 2 * image_width / 3:
        return "center"
    return "right"


def extract_predictions(result, conf_thresh):
    pred = result.pred_instances
    boxes = pred.bboxes.cpu().numpy()
    scores = pred.scores.cpu().numpy()
    labels = pred.labels.cpu().numpy()
    keep = scores >= conf_thresh
    return boxes[keep], scores[keep], labels[keep]


def draw_predictions(frame, boxes, scores, labels, class_names):
    h, w = frame.shape[:2]
    direction_counts = {"left": 0, "center": 0, "right": 0}
    for box, score, label in zip(boxes, scores, labels):
        direction = get_direction(box, w)
        direction_counts[direction] += 1
        x1, y1, x2, y2 = box.astype(int)
        class_name = class_names[label] if class_names is not None and label < len(class_names) else str(label)
        text = f"{class_name} {score:.2f} {direction}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, text, (x1, max(y1 - 10, 20)), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (0, 255, 0), 2)
    return frame, direction_counts


def main():
    args = parse_args()
    os.makedirs(args.output, exist_ok=True)
    annotated_dir = os.path.join(args.output, "annotated_images")
    os.makedirs(annotated_dir, exist_ok=True)

    print("Loading model...")
    model = init_detector(args.config, args.checkpoint, device=args.device)
    class_names = model.dataset_meta.get("classes", None)
    print("Model loaded.")

    image_files = [os.path.join(args.input, f) for f in sorted(os.listdir(args.input)) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    if not image_files:
        raise RuntimeError(f"No images found in {args.input}")

    for img_path in image_files[:5]:
        img = cv2.imread(img_path)
        if img is not None:
            _ = inference_detector(model, img)
    if args.device.startswith("cuda") and torch.cuda.is_available():
        torch.cuda.synchronize()

    latencies = []
    total_detections = 0
    images_with_detection = 0
    images_without_detection = 0
    direction_counts = {"left": 0, "center": 0, "right": 0}
    per_image_rows = []

    for img_path in tqdm(image_files, desc="Testing validation images"):
        frame = cv2.imread(img_path)
        if frame is None:
            print(f"Skipping unreadable image: {img_path}")
            continue

        start = time.time()
        result = inference_detector(model, frame)
        if args.device.startswith("cuda") and torch.cuda.is_available():
            torch.cuda.synchronize()
        latency = time.time() - start
        latencies.append(latency)

        boxes, scores, labels = extract_predictions(result, args.conf)
        num_detections = len(boxes)
        total_detections += num_detections
        images_with_detection += int(num_detections > 0)
        images_without_detection += int(num_detections == 0)

        frame, frame_direction_counts = draw_predictions(frame, boxes, scores, labels, class_names)
        for key in direction_counts:
            direction_counts[key] += frame_direction_counts[key]

        cv2.imwrite(os.path.join(annotated_dir, os.path.basename(img_path)), frame)
        per_image_rows.append({
            "image": os.path.basename(img_path),
            "latency_sec": round(latency, 6),
            "detections": num_detections,
            "left_detections": frame_direction_counts["left"],
            "center_detections": frame_direction_counts["center"],
            "right_detections": frame_direction_counts["right"],
        })

    avg_latency = sum(latencies) / len(latencies)
    fps = 1 / avg_latency
    missed_frame_rate_estimate = images_without_detection / len(latencies)

    metrics = {
        "input_type": "validation_image_folder",
        "images_tested": len(latencies),
        "total_detections": total_detections,
        "images_with_detection": images_with_detection,
        "images_without_detection": images_without_detection,
        "missed_frame_rate_estimate": round(missed_frame_rate_estimate, 4),
        "average_latency_sec_per_image": round(avg_latency, 4),
        "fps": round(fps, 2),
        "left_detections": direction_counts["left"],
        "center_detections": direction_counts["center"],
        "right_detections": direction_counts["right"],
        "confidence_threshold": args.conf,
    }

    save_summary_csv(args.output, metrics)
    save_per_image_csv(args.output, per_image_rows)
    print("\n===== CUSTOM VALIDATION IMAGE RESULTS =====")
    for k, v in metrics.items():
        print(f"{k}: {v}")


if __name__ == "__main__":
    main()
```
